```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, b, c, d, e = -12, -18, 5, 10, -30
```

```python
def func(x):
    return a*x**4*np.sin(np.cos(x)) + b*x**3 + c*x**2 + d*x + e
```

```python
x = np.arange(-50, 50, 0.01)
```

```python
def take_roots(a, b, c, d, e):
    return np.roots([a, b, c, d, e])
```

```python
roots = take_roots(a, b, c, d, e)
print(roots)
```

```python
plt.plot(x, func(x), 'b')
plt.plot(roots[0], 0, 'rx')
plt.plot(roots[1], 0, 'rx')
plt.plot(roots[2], 0, 'rx')
plt.plot(roots[3], 0, 'rx')
plt.grid()
```
